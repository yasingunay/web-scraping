```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

```

```python
url = "https://www.ideaconsult.be/en/about-us/team"

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
response = requests.get(url, headers=headers)
result = response.text

print(result)
```

```python
soup = BeautifulSoup(result, "html.parser")

team = []

members = soup.find_all("div", class_="idea-member")
# print(members)

for number, member in enumerate(members, start=1):
    name = member.find("div", class_="idea-member-name").text.strip()
    role = member.find("div", class_="idea-member-function").text.strip()
    phone = member.find("div", class_="idea-member-contact-phone").text.strip()
    email = member.find("div", class_="idea-member-contact-email").text.strip()
    team.append({"number": number, "name": name, "role": role, "phone": phone, "email": email})


    
```

```python


df = pd.DataFrame(team)
df = df[['number', 'name', 'role', 'phone', 'email']]
df.to_excel("team.xlsx", index=False)
```
